```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn

#Lendo o arquivo CSV

df = pd.read_csv('HIST_PAINEL_COVIDBR_28out2020.csv', sep=';')

df.shape
```

```python
#Filtrando o arquivo para ter somente as linhas datadas do dia 28 de outubro

planilha_recente = df.loc[df['data'] == '2020-10-28']
planilha_recente.shape
```

```python
#Filtrando o arquivo para ter somente as informações dos estados
#OBS: o .iloc[1:] é por que a linha zero é referente aos dados do Brasil e possui municipio null, mas não é um estado.

estados = planilha_recente.loc[planilha_recente['municipio'].isnull()].iloc[1:]
estados
```

```python
estados.shape
```

```python
#Eliminando as informações duplicadas da planilha filtrada

Estados_definitivo = estados.drop_duplicates(subset='estado')
Estados_definitivo
```

```python
Estados_definitivo.shape
```

```python
#Salvando a planilha filtrada como um novo arquivo csv que será usado para a construção dos gráficos e rankings

Estados_definitivo.to_csv('Planilha Limpa.csv')
```
